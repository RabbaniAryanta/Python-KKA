```python
# 1. Variabel dan Tipe Data
angka = 10          # int
desimal = 3.14      # float
teks = "Halo Python" # string
daftar = [1, 2, 3]   # list

print(type(angka))
print(type(desimal))
print(type(teks))
print(type(daftar))
```

```python
# 2. List dan Manipulasi
belanja = ["beras", "minyak", "telur"]

def tambah_item(item):
    belanja.append(item)

tambah_item("gula")
tambah_item("kopi")

for barang in belanja:
    print(barang)
```

```python
# 3. Dictionary
harga = {
    "beras": 12000,
    "minyak": 17000,
    "telur": 24000,
    "gula": 15000,
    "kopi": 20000
}

total = sum(harga.values())
print("Total harga belanjaan:", total)
```

```python
# 4. Fungsi
def hitung_persegi_panjang(panjang, lebar):
    luas = panjang * lebar
    keliling = 2 * (panjang + lebar)
    return luas, keliling

luas, keliling = hitung_persegi_panjang(5, 3)
print("Luas:", luas)
print("Keliling:", keliling)

```

```python
# 5. Percabangan
usia = int(input("Masukkan usia: "))

if 0 <= usia <= 13:
    kategori = "Anak"
elif 14 <= usia <= 24:
    kategori = "Remaja"
elif 25 <= usia <= 49:
    kategori = "Dewasa"
else:
    kategori = "Lansia"

print("Kategori usia:", kategori)
```
